```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk import pos_tag
from nltk.stem import WordNetLemmatizer,PorterStemmer
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
```

```python
document="The quick brown fox jumped over the lazy dog. The dog didn't seem to care."
```

```python
tokens=word_tokenize(document)
print(f"Tokens: {tokens}\n")
```

```python
pos_tags=nltk.pos_tag(tokens)
print(f"POS tags: {pos_tags}\n")
```

```python
stop_words=set(stopwords.words('english'))
filtered_tokens=[word for word in tokens
 if word.casefold()not in stop_words]
print(f"Filtered tokens: {filtered_tokens}\n")
```

```python
lemmatizer = WordNetLemmatizer()

token = word_tokenize(document)

lemmatized_output = [lemmatizer.lemmatize(word) for word in token]
print(" ".join(lemmatized_output))
```

```python
stemmer = PorterStemmer()

token = word_tokenize(document)

stemmed = [stemmer.stem(word) for word in token]
print(" ".join(stemmed))
```
